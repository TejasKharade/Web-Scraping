```python
import requests
from bs4 import BeautifulSoup
```

```python
a=requests.get('https://www.gutenberg.org/ebooks/categories')

with open(f"folder/book_page.html",'w',encoding='utf-8') as f:
    f.write(a.text)
    print("Page Loaded Successfully")
```

```python
with open(f"folder/book_page.html", encoding='utf-8') as f:
    content=f.read()
    soup=BeautifulSoup(content,'html.parser')
```

```python
print(soup)
```

```python
bookshelf_div=soup.find('div', class_='bookshelves')
print(bookshelf_div)
```

```python
main_link="https://www.gutenberg.org/ebooks"

list=[]

for section in bookshelf_div.find_all(['h2','ul']):
    if section.name=='h2':
        main_category=section.text.strip()
    if section.name=='ul':
        for li in section.find_all('li'):
            sub_category=li.find('a').text.strip()
            url= main_link+li.find('a')['href']
            list.append([main_category,sub_category,url])

print(list)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(list, columns=['Category','Sub-Category','URL'])
```
